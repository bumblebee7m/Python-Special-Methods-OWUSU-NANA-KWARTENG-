# 1. Constructors in Python (__init__())

## Definition

The `__init__()` method is a special method (constructor) that is automatically called whenever an object is created from a class.

It is mainly used to initialize the attributes of an object, ensuring that every object starts with the required values.

```python
# Example 1

class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

s1 = Student("Matthew", 30)

print(s1.name)
print(s1.age)

# Example 2

class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author

    def describe(self):
        print(f"{self.title} by {self.author}")

b1 = Book("OOP", "Dr. Cobbinah")
b1.describe()
```

# 2. String Representation (__str__())

## Definition

The `__str__()` method provides a human-readable representation of an object.

Whenever the `print()` function is used on an object, Python automatically calls the `__str__()` method.

Without this method, Python displays the object's memory location, which is not user-friendly.


```python
class Student:

    def __init__(self, name):
        self.name = name

    def __str__(self):
        return f"Student: {self.name}"

s = Student("Matthew")

print(s)
```

# 3. Formal Representation (__repr__())

## Definition

The `__repr__()` method provides an official or developer-oriented representation of an object.

It is mainly used for debugging and logging.

If the `__str__()` method is not defined, Python automatically uses the `__repr__()` method.


```python
# Example

class Student:
    def __init__(self, name):
        self.name = name

    def __repr__(self):
        return f"Student('{self.name}')"

s = Student("Daniella")

print(repr(s))
```

# 4. Length Method (__len__())
Definition

The __len__() special method allows a custom object to define its own length. It enables Python's built-in len() function to work with user-defined classes.

```python
# Example

class Playlist:
    def __init__(self, songs):
        self.songs = songs

    def __len__(self):
        return len(self.songs)

p = Playlist(["Song1", "Song2", "Song3"])

print(len(p))
```

# 5. Equality Comparison (__eq__())
Definition

The __eq__() method defines how two objects should be compared for equality. Instead of comparing memory addresses, it compares the contents of the objects.

```python
# Example

class Student:
    def __init__(self, student_id):
        self.student_id = student_id

    def __eq__(self, other):
        return self.student_id == other.student_id

s1 = Student(101)
s2 = Student(101)

print(s1 == s2)
```

# 6. Addition (__add__())
Definition

The __add__() method overloads the addition (+) operator for custom objects. It allows objects to be added together in a meaningful way.

```python
# Example

class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(
            self.x + other.x,
            self.y + other.y
        )

    def __str__(self):
        return f"({self.x}, {self.y})"

v1 = Vector(2, 3)
v2 = Vector(4, 5)

print(v1 + v2)
```

# 7. Subtraction (__sub__())
Definition

The __sub__() method overloads the subtraction (-) operator for objects. It defines how subtraction should behave between two custom objects.

```python
# Example

class Vector:
    def __init__(self, x):
        self.x = x

    def __sub__(self, other):
        return Vector(self.x - other.x)

    def __str__(self):
        return str(self.x)

v1 = Vector(10)
v2 = Vector(3)

print(v1 - v2)
```


# 8. Multiplication (__mul__())
Definition

The __mul__() method defines how the multiplication (*) operator works with custom objects. It makes user-defined classes behave like built-in numeric types.

```python
# Example

class Number:
    def __init__(self, value):
        self.value = value

    def __mul__(self, other):
        return Number(self.value * other.value)

    def __str__(self):
        return str(self.value)

n1 = Number(5)
n2 = Number(4)

print(n1 * n2)
```

# 9. Item Access (__getitem__())
Definition

The __getitem__() method allows objects to support indexing like lists and tuples. It is called whenever an object is accessed using square brackets.

```python
# Example

class Grades:
    def __init__(self):
        self.data = [85, 90, 78]

    def __getitem__(self, index):
        return self.data[index]

g = Grades()

print(g[0])
print(g[1])
```

# 10. Item Assignment (__setitem__())
Definition

The __setitem__() method allows values inside an object to be modified using indexing. It behaves similarly to assigning values in a list or dictionary.

```python
# Example

class Grades:
    def __init__(self):
        self.data = [85, 90, 78]

    def __setitem__(self, index, value):
        self.data[index] = value

g = Grades()

g[1] = 100

print(g.data)
```

# 11. Membership Testing (__contains__())
Definition

The __contains__() method defines how the in operator behaves for custom objects. It checks whether a particular item exists in the object.

```python
# Example

class Course:
    def __init__(self):
        self.students = ["Ella", "Sandra", "Joe"]

    def __contains__(self, student):
        return student in self.students

c = Course()

print("Joe" in c)
print("Gifty" in c)
```

# 12. Callable Objects (__call__())
Definition

The __call__() method allows an object to behave like a function. Once implemented, the object can be called using parentheses.

```python
# Example

class Multiplier:
    def __init__(self, factor):
        self.factor = factor

    def __call__(self, number):
        return number * self.factor

double = Multiplier(2)

print(double(10))
```

# 13. Iteration Support (__iter__() and __next__())
Definition

The __iter__() and __next__() methods implement Python's iterator protocol. They allow objects to be used in for loops and other iteration operations.

```python
# Example

class Counter:
    def __init__(self, limit):
        self.limit = limit
        self.current = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.current > self.limit:
            raise StopIteration

        value = self.current
        self.current += 1
        return value

for num in Counter(5):
    print(num)
```

# 14. Boolean Evaluation (__bool__())
Definition

The __bool__() method determines whether an object is considered True or False. It is automatically used in Boolean expressions and conditional statements.

```python
# Example

class BankAccount:
    def __init__(self, balance):
        self.balance = balance

    def __bool__(self):
        return self.balance > 0

account = BankAccount(500)

if account:
    print("Account is active")
```

# 15. Destructor (__del__())
Definition

The __del__() method is called when an object is about to be destroyed. It is mainly used for resource cleanup, although modern Python often prefers context managers.

```python
# Example

class Resource:
    def __init__(self):
        print("Resource allocated")

    def __del__(self):
        print("Resource released")

r = Resource()
```
